```python
from pygelbooru import Gelbooru
from dotenv import load_dotenv
import os
import json
load_dotenv()

API_KEY = os.getenv("API_KEY")
USER_ID = os.getenv("USER_ID")
EXCLUDE_TAGS = os.getenv("EXCL_TAGS")


gelbooru = Gelbooru(API_KEY, USER_ID)
```

```python
async def download_char_images(_tags: list, _exclude_tags: list, _char_name: str, _limit: int = 100):
    print(f"Downloading {_char_name} images")
    import requests
    results = await gelbooru.search_posts(tags=_tags, exclude_tags=_exclude_tags, limit=_limit)

    result_urls = [post.file_url for post in results]

    for post in results:
        print(post.file_url)

    # Download images
    os.makedirs(f"data/GI/{_char_name}", exist_ok=True)

    for i, url in enumerate(result_urls):
        r = requests

        with open(f"data/GI/{_char_name}/{i}.jpg", "wb") as f:
            f.write(r.get(url).content)
        print(f"Downloaded {i}.jpg")



```

```python
import os
async def add_images_to_folder(path, tags, char_name, exclude_tags, num_images):

  final_path = path + char_name
  print(final_path)
  i = len(os.listdir(final_path))
  print(f'there are {i} images in the folder')

  # there are images 0 to i-1 already in the folder
  
  for idx in range(i, i+num_images):
    result = await gelbooru.random_post(tags=tags, exclude_tags=exclude_tags)
    url = result.file_url
    import requests
    r = requests


    with open(f"{path}{char_name}/{idx}.jpg", "wb") as f:
          f.write(r.get(url).content)

    print(f"Downloaded {idx}.jpg")

test_dir = "data/GI/test/"

# await add_images_to_folder(test_dir, "Ganyu", EXCLUDE_TAGS, 32)
    
tags = ["ganyu_(genshin_impact)", "1girl", "solo"]
exclude_tags = EXCLUDE_TAGS.split(" ")
char_name = "Ganyu"

await add_images_to_folder(test_dir, tags, char_name, exclude_tags, 2)

# result = await gelbooru.random_post(tags = tags, exclude_tags = exclude_tags)
# print(result.file_url)
```

```python
train_dir = "data/GI/train/"

async def add_to_train_folder(char_name, num_images):
  chname = char_name.lower()
  tags = [f"{chname}_(genshin_impact)", "1girl", "solo"]
  exclude_tags = EXCLUDE_TAGS.split(" ")
  await add_images_to_folder(train_dir, tags, char_name, exclude_tags, num_images)
```

```python
async def batch_add_to_train_folder(char_names, num_images):
  for name in char_names:
    await add_to_train_folder(name, num_images)
```

```python
char_names = ["Fischl", "Nahida", "Hu_Tao", "Furina"]

await batch_add_to_train_folder(char_names, 184)
```
